<p style="font-size:25px">Data Visualization</p>


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
try:
    df = pd.read_csv('google_trends_data.csv')
except FileNotFoundError:
    print("Error: google_trends_data.csv not found. Please upload the file.")
    exit()
if df.empty:
    print("Error: google_trends_data.csv is empty.")
    exit()
try:
    category_col = df.columns[0]
    value_col = df.columns[1]
    df = df.rename(columns={category_col: 'Category', value_col: 'Value'})
except IndexError:
    print("Error: google_trends_data.csv does not have enough columns for plotting.")
    exit()


# 1. Vertical Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value', data=df)
plt.title('Vertical Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Horizontal Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Value', y='Category', data=df)
plt.title('Horizontal Bar Chart')
plt.xlabel('Value')
plt.ylabel('Category')
plt.tight_layout()
plt.show()

if len(df.columns) < 3:
    print("Adding a 'Group' column for demonstration of Grouped/Stacked charts.")
    # Create a dummy 'Group' column
    import numpy as np
    df['Group'] = np.random.choice(['Group A', 'Group B'], size=len(df))

# Grouped Bar Chart
if 'Group' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Category', y='Value', hue='Group', data=df)
    plt.title('Grouped Bar Chart')
    plt.xlabel('Category')
    plt.ylabel('Value')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Group')
    plt.tight_layout()
    plt.show()
else:
    print("Skipping Grouped Bar Chart: 'Group' column not found or created.")

# Bar Plot with Color by Category
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value', hue='Category', data=df, palette='viridis', legend=False)
plt.title('Bar Plot with Color by Category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
